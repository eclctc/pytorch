###2. Create a random tensor with the shape (7,7)

```python
import torch
tensor1 = torch.rand(size=(7,7))
tensor1, tensor1.shape
```

###3. Perform a matrix mulitplication on the tensor from 2 with another random tensor with shape (1,7)

```python
tensor2 = torch.rand(size=(1,7))
tensor2, tensor2.shape
```

```python
tensor3 = torch.matmul(tensor1, tensor2.T)
tensor3, tensor3.shape
```

###4. Set the random seed to 0 and do 2 & 3 over again

```python
torch.manual_seed(0)
tensor_A = torch.rand(7, 7)

torch.manual_seed(0)
tensor_B = torch.rand(1, 7)

tensor_c = torch.matmul(tensor_A, tensor_B.T)

tensor_c, tensor_c.shape
```
