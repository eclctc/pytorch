## Random Tensors

Why random tensors?

Random tensors are important because they way many neural networks learn is that they start with tensors full of random numbers and the adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

Torch random tensors - https://docs.pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(1, 3, 4)
random_tensor
```

```python
random_tensor.ndim
```

### Tensor datatypes

**Note** Tensor datatypes is one of the 3 errors you'll run into with PyTorch and deep learning
1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device


```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               # These are very important when you're defining tensors
                               dtype=None,          # What datatype is the tensor (eg float32, float16)
                               device=None,         # What device is your tensor on
                               requires_grad=False)  # Whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor * float_32_tensor  # This works because they're both floats
```

```python
int_32_tensor = torch.tensor([3, 6, 8], dtype=torch.long)
int_32_tensor * float_32_tensor
```

## Matrix multiplication

Two main way of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

More information on multiplying matricies - https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

2. The resulting matrix has the shape of the **outer dimensions**
* `(2,3) @ (3,2)` -> The resulting shape in (2,2)
* `(3,2) @ (2,3)` -> The resulting shape in (3,3)

```python
torch.matmul(torch.rand(3,2), torch.rand(3,2))
```

```python
torch.matmul(torch.rand(3,2), torch.rand(2,3))
```

## One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,  2],
                         [3,  4],
                         [5,  6]])
tensor_B = torch.tensor([[7,  10],
                         [8,  11],
                         [9,  12]])
# torch.mm(tensor_A, tensor_B) torch.mm is the same thing as matmul (it's as alias for writing less code)
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues, we can manipulte the shape of one of our tensors using a **transpose**

A **transpose** switches the axes of dimensions of a given tensor.

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Ouput:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```
