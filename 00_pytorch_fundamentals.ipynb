# Introduction to tensors

## Creating tensors

PyTorch tensors are created using `torch.Tensor` -= https://docs.pytorch.org/docs/stable/tensors.html

```python
# scalar
scalar = torch.tensor(7)
scalar
```

```python
scalar.ndim
```

```python
scalar.item()
```

```python
# Vector
vector = torch.tensor([7,7])
vector
```

```python
# You can think of dimensions as however many square brackets there are
vector.ndim
```

```python
# You can think of the shape as however many pairs are in the tensor
vector.shape
```

```python
# MATRIX
MATRIX = torch.tensor([[7,8],
                       [9,10]])
MATRIX
```

```python
MATRIX.ndim
```

```python
MATRIX[1]
```

```python
MATRIX.shape
```

```python
# Tensor
TENSOR = torch.tensor([[[1,2,3],
                        [3, 6, 9],
                        [2, 4, 5]]])
TENSOR
```

```python
TENSOR.ndim
```

```python
TENSOR.shape
# This means that we have one tensor that is 3 x 3 in shape
```
