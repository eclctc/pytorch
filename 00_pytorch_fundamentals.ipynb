## Random Tensors

Why random tensors?

Random tensors are important because they way many neural networks learn is that they start with tensors full of random numbers and the adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

Torch random tensors - https://docs.pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(1, 3, 4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a raondom tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 224)) #height, width, color channels (R, G, B)
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

# Create a range of tensors and tensors-like

```python
# Use torch.range() and get deprecated message, use torch.arrange()
torch.arange(0, 10)
one_to_ten = torch.arange(start=1, end=11, step=1)
one_to_ten
```

### Tensor datatypes

**Note** Tensor datatypes is one of the 3 errors you'll run into with PyTorch and deep learning
1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device


```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               # These are very important when you're defining tensors
                               dtype=None,          # What datatype is the tensor (eg float32, float16)
                               device=None,         # What device is your tensor on
                               requires_grad=None)  # Whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```
